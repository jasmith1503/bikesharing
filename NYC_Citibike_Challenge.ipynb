```python
import pandas as pd
import os
from datetime import datetime
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
citibike_dataset = os.path.join('./201908-citibike-tripdata.csv')
citibike_df = pd.read_csv(citibike_dataset)
citibike_df.sample(10)
```

```python
# 2. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# 3-a Make a copy of the column "tripduration" and add it to the end of the DF and name it tripduration_int
citibike_df['tripduration_int'] = citibike_df['tripduration']
citibike_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
citibike_df['tripduration'] = pd.to_datetime(citibike_df['tripduration'], unit='s')
# citibike_df['tripduration'] = pd.to_datetime(citibike_df['tripduration'],'s')
```

```python
# 4. Check the datatypes of your columns. 
citibike_df.dtypes
```

```python
# View sample of the updated DF
citibike_df.sample(7)
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
output_data_filename = "NYC_CityBike_data_cleaned.csv"
citibike_df.to_csv(output_data_filename, index=False)
```
